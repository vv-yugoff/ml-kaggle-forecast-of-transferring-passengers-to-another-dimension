```python
import pandas as pd
from sklearn.linear_model import LinearRegression

train_data_path = pd.read_csv('train.csv')
test_data_path = pd.read_csv('test.csv')

X_train = train_data_path.drop(
    'Transported',
    axis=1
)
y_train = train_data_path['Transported']

X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(test_data_path)

X_train_dropna = X_train_encoded.dropna(axis=1)
X_test_dropna = X_test_encoded.dropna(axis=1)

X_train_dropna, X_test_dropna = X_train_dropna.align(
    X_test_dropna, 
    join='inner', 
    axis=1
)

model = LinearRegression()
model.fit(X_train_dropna, y_train)

y_prediction = model.predict(X_test_dropna)

id_column = test_data_path['PassengerId']
predictions = pd.DataFrame({
    'PassengerId': id_column, 
    'Transported': y_prediction
})

threshold = 0.5
predictions['Transported'] = predictions['Transported'].apply(lambda x: True if x > threshold else False)

predictions.to_csv(
    'predictions.csv', index = False
)
```
